# 力と運動の法則

## 運動の第二法則/ 運動方程式

加速度: $a$, 力:$F$, 質量: $m$ とすると, この三つは次の式[運動方程式]で取りまとめることができる.

>> $\qquad m \cdot a = F$

この運動方程式を加速度を中心にまとめると $a$ を
\begin{align*}
a &= \dfrac{F}{M},&\\
\end{align*}
これを比例記号 $\propto$ を用いて分けると
\begin{align*}
a &\propto F&\\
a &\propto \dfrac{1}{M}&\\
\end{align*}
これを言葉で表すと
- 加速度は力に比例する
- 加速度は質量に反比例する
ということができます. 以下ではこれを順番に確認していきます.

加速度: $a$, 力: $F$

ここで示したいことは
\begin{align*}
a \propto F
\end{align*}
です. これは
> 加速度 $a$ は力 $F$ に比例している

## 運動方程式 $ma = F$

実験でわかることは
\begin{align*}
 a &\propto F&\\
 a &\propto \dfrac{1}{m}&
\end{align*}
つまり
\begin{align*}
 a &\propto \dfrac{F}{m}&
\end{align*}
です. この比例関係を等式に変換するには
比例定数 $c$ が必要です.
\begin{align*}
 a &= c\  \dfrac{F}{m}&
\end{align*}
この比例定数 $c$ を他の実験とつじつまが合うように($m={\rm 1\ kg}, a={\rm 1 \ m/s^2}$) のときに, $c=1$ とし
\begin{align*}
 a &= 1\  \dfrac{F}{m}&\\
 m \cdot a & = F&\\
\end{align*}
がえられます. これを

>> **運動方程式** $\qquad m \cdot a = F$

と呼ばれます. 言葉でまとめると

> 質量 $m$ の物体に力 $F$ が働くと, $F$ と同じ向きに 加速度が生じる. また, その加速度 $a$ の大きさは, 力 $F$ の大きさに比例し, 質量 $m$ に反比例する.


### 力の単位の定義

運動方程式から $m={\rm 1\ kg}, a={\rm 1 \ m/s^2}$
\begin{align*}
 F &= m \cdot a &\\
 &= {\rm 1\ kg} \cdot {\rm 1 \ m/s^2} = {\rm 1 \ kg\ m/s^2}&\\
 &= {\rm 1 \ N}
\end{align*}

>> 力の単位は ${\rm 1 \ N}$ で,  ${\rm 1 \ kg\ m/s^2}$ である.  

ということができます.
